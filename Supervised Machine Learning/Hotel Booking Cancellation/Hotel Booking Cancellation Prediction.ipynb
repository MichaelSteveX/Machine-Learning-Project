#### 05 Feature Scalling

````python
# Select feature for x and y  
x = df2[[ 'hotel',
    'arrival_date_year',
    'arrival_date_month',
    'country',
    'reservation_status',
    'assigned_room_type',
    'customer_type']]
        
y = df2['is_canceled']
````

````python
x.head()
````

````python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=52)
````

````python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit_transform(x)
````

#### 06 Model Apply

````python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier, VotingClassifier
````

````python
# apply logestic regression 
lg = LogisticRegression(max_iter=1000)
lg.fit(x_train, y_train)
lg.score(x_test,y_test)*100 , lg.score(x_train,y_train)*100
````

````python
# apply decision Tree 
dc = DecisionTreeClassifier(max_depth=6)
dc.fit(x_train,y_train)
dc.score(x_test,y_test)*100 , dc.score(x_train,y_train)*100
````

````python
rf = RandomForestClassifier(n_estimators=100,random_state=42)
rf.fit(x_train, y_train)
rf.score(x_test,y_test)*100 , rf.score(x_train,y_train)*100
````

#### 07 Confusion Metrix

````python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
````

````python
# Compute confusion matrix
cf = confusion_matrix(y_test, dc.predict(x_test))
cf
````

````python
precision_score(y_test, dc.predict(x_test))*100
````

````python
recall_score(y_test, dc.predict(x_test))*100
````

````python
f1_score(y_test, dc.predict(x_test))*100
````

#### 08 Check Errors

````python
from sklearn.metrics import mean_squared_error, mean_absolute_error

mean_squared_error(y_test, dc.predict(x_test)) , mean_absolute_error(y_test, dc.predict(x_test))
````
